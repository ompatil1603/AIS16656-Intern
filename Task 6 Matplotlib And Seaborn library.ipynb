Matplotlib
Matplotlib is a versatile plotting library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt

# Example data
x = [5,7,8,9,12,15]
y = [4,6,9,12,17,22]

# Create a line plot
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Create a scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Create a bar plot
plt.bar(x, y)
plt.title('Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Example data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# Create a histogram
plt.hist(data, bins=4)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

Seaborn
Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns

# Example data
data = sns.load_dataset('fmri')

# Create a line plot
sns.lineplot(data=data, x='timepoint', y='signal', hue='region')
plt.title('Line Plot')
plt.show()
```

```python
# Example data
data = sns.load_dataset('tips')

# Create a scatter plot
sns.scatterplot(data=data, x='total_bill', y='tip', hue='day')
plt.title('Scatter Plot')
plt.show()
```

```python
# Create a bar plot
sns.barplot(data=data, x='day', y='total_bill')
plt.title('Bar Plot')
plt.show()
```

```python
# Create a histogram
sns.histplot(data=data, x='total_bill', bins=10)
plt.title('Histogram')
plt.show()
```

```python
# Create a box plot
sns.boxplot(data=data, x='day', y='total_bill')
plt.title('Box Plot')
plt.show()
```

```python
# Create a violin plot
sns.violinplot(data=data, x='day', y='total_bill')
plt.title('Violin Plot')
plt.show()
```

```python
# Example data
x = range(1, 6)
y1 = [1, 2, 4, 8, 16]
y2 = [1, 3, 6, 10, 15]

# Create an area plot
plt.fill_between(x, y1, color="skyblue", alpha=0.4)
plt.fill_between(x, y2, color="olive", alpha=0.4)
plt.title('Area Plot')
plt.show()
```

```python
import seaborn as sns

# Example data
data = sns.load_dataset('tips')

# Create a point plot
sns.pointplot(x='day', y='total_bill', data=data)
plt.title('Point Plot')
plt.show()
```

```python
import seaborn as sns

# Example data
data = sns.load_dataset('tips')

# Create a count plot
sns.countplot(x='day', data=data)
plt.title('Count Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')

# Matplotlib Line Plot
plt.figure(figsize=(15, 10))

# Subplot 1: Line Plot
plt.subplot(3, 3, 1)
plt.plot(tips['total_bill'], tips['tip'], color='blue')
plt.title('Line Plot')

# Subplot 2: Scatter Plot
plt.subplot(3, 3, 2)
plt.scatter(tips['total_bill'], tips['tip'], color='red')
plt.title('Scatter Plot')

# Subplot 3: Histogram
plt.subplot(3, 3, 3)
plt.hist(tips['total_bill'], bins=10, color='green')
plt.title('Histogram')

# Subplot 4: Box Plot
plt.subplot(3, 3, 4)
sns.boxplot(x='day', y='total_bill', data=tips, ax=plt.gca())
plt.title('Box Plot')

# Subplot 5: Violin Plot
plt.subplot(3, 3, 5)
sns.violinplot(x='day', y='total_bill', data=tips, ax=plt.gca())
plt.title('Violin Plot')

plt.tight_layout()
plt.show()
```
